```python
# BBQ restaurants in Houston
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
from yelpapi import YelpAPI
from config import yelp_API
```

```python
url='https://api.yelp.com/v3/businesses/search'
```

```python
# BBQ restaurants in Houston
headers = {'Authorization':'Bearer '+ yelp_API}
params_set1 = {'location': 'Houston, TX',
               'limit': 50,
               'offset':0,
          'categories': 'bbq',
         'term': 'restaurant'}
response = requests.get(url, headers=headers, params=params_set1)
response_json=response.json()
res_lat=[]
res_long=[]
res_name=[]
res_rating=[]
res_vote=[]
for respo in response_json:
    nam=response_json['businesses']
    for na in nam:
        try:
            name= na['name']
            rating= na['rating']
            vote=na['review_count']
            lat=na['coordinates']['latitude']
            long=na['coordinates']['longitude']
            res_name.append(name)
            res_rating.append(rating)
            res_vote.append(vote)
            res_lat.append(lat)
            res_long.append(long)
        except:
            print('NA')
df_yelp_houston_bbq_set1=pd.DataFrame({'Name': res_name,
                              'Latitude':res_lat,
                              'Longitude': res_long,
                              'Votes': res_vote,
                              'Rating': res_rating})
df_yelp_houston_bbq_set1
```

```python
# Drop duplicates:
df_yelp_houston_bbq_all=df_yelp_houston_bbq_set1.drop_duplicates(keep='first')
df_yelp_houston_bbq_all
```

```python
#### Save raw data to csv
df_yelp_houston_bbq_all.to_csv('data\df_yelp_houston_bbq_all_NN.csv',index=False)
```
