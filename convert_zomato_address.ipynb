```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Google API Key
from config import gkey
```

```python
## read raw data
zomato_mexican_houston= pd.read_csv('C:\\workspace\\Resources\\test_df_restaurant_houston_raw_NN.csv')
zomato_mexican_houston
```

```python
# convert data
longitude=[]
latitude=[]

address = zomato_mexican_houston["Address"].tolist()

for add in address:
    
    params = {
    "address": add,
    "key": gkey}
    
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    
    response_json = requests.get(base_url, params=params).json()
    results= response_json['results']
    for re in results:
        lat= re['geometry']['location']['lat']
        long= re['geometry']['location']['lng']
        latitude.append(lat)
        longitude.append(long)
```

```python
# Add new columns
zomato_mexican_houston['Longitude'] = longitude
zomato_mexican_houston['Latitude']= latitude
zomato_mexican_houston[['Name', 'Longitude', 'Latitude', 'Votes','Rating']]
zomato_mexican_houston
```

```python
#### Save converted data to csv
zomato_mexican_houston.to_csv('zomato_mexican_houston_NN.csv',index=False)
```
