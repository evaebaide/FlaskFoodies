```python
# scrape 
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup as bs
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
import requests
import pandas as pd
import time
```

<h1># Windows Users</h1>


```python
#Set the chromedriver path
executable_path = {"executable_path": "/users/evaeb/bin/chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
# Scrap Middle Eastern restaurants in New York
# url = 'https://www.zomato.com/new-york-city/restaurants/middle-eastern'
n = 1
restaurant_name=[]
restaurant_address=[]
votes=[]
ratings=[]
costs=[]

for x in range(1, 30):
    url = f'https://www.zomato.com/new-york-city/restaurants/mediterranean?page={n}'
    browser.visit(url)

    html = browser.html
    soup = bs(html, 'html.parser')
    restaurant_info = soup.find_all(class_='content')
    
    for restaurant in restaurant_info:
        try:
            rest_name= restaurant.find('a', class_='result-title').text     
            rest_address= restaurant.find(class_='search-result-address').text
            vote= restaurant.find('span').text
            rating= restaurant.find(class_='rating-popup').text
            cost= restaurant.find(class_="res-cost").text
            
                  
            restaurant_name.append(rest_name)
            restaurant_address.append(rest_address)
            votes.append(vote)
            ratings.append(rating)
            costs.append(cost)
            
        except:
            print("Not Available")
              
        
#         print(rest_name)
#         print(rest_address)
#         print(vote)
#         print(rating)
        n += 1
```
