```python
# Scrap Mexican food information in Houston
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import pandas as pd
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Scrap Mexican food information in Houston
url = 'https://www.zomato.com/houston/restaurants/mexican'
browser.visit(url)
restaurant_name=[]
restaurant_address=[]
votes=[]
ratings=[]

for x in range(1, 30):
    html = browser.html
    soup = bs(html, 'html.parser')
    restaurant_info = soup.find_all(class_='content')
    for restaurant in restaurant_info:
        try: 
            rest_name= restaurant.find('a', class_='result-title').text
            rest_address= restaurant.find(class_='search-result-address').text
            vote= restaurant.find('span').text
            rating= restaurant.find(class_='rating-popup').text
        except:
            print("Not Available")
        restaurant_name.append(rest_name)
        restaurant_address.append(rest_address)
        votes.append(vote)
        ratings.append(rating)
        
#         print(rest_name)
#         print(rest_address)
#         print(vote)
#         print(rating)
        try:
            browser.click_link_by_partial_text('Next Page')
        except:
            print("Scraping Complete")
        
```
